feature eng

```python
df['Mark_cat']=df['gmat'].apply(lambda x:'High' if x>=700 else 'Average' if x<=570 else 'Below Average')
```

```python
df['exp_cat'] = df['work_exp'].apply(lambda x:'Senior' if x>=7 else 'Intermediate' if x<=4 else 'Entry Level')
```

```python
df['normalized_gpa'] = (df['gpa'] - df['gpa'].min()) / (df['gpa'].max() - df['gpa'].min())
df['normalized_gmat'] = (df['gmat'] - df['gmat'].min()) / (df['gmat'].max() - df['gmat'].min())
df['combined_normalized_score'] = df['normalized_gpa'] + df['normalized_gmat']
```

```python
top_10_scores = df['combined_normalized_score'].sort_values(ascending=False).head(10)
top_10_scores
```

Average GMAT Score by Major: This will help reveal if students from certain majors tend to score higher on the GMAT



```python
average_gmat_by_major=df[['major','gmat']].groupby('major').mean()
print("Average GMAT score by major:\n", average_gmat_by_major)
```

Work Experience vs. GMAT Score: We can assess whether candidates with more work experience score higher on the GMAT on average.

```python
correlation_work_exp_gmat = df[['exp_cat','gmat']].groupby('exp_cat').mean()
print(f"Correlation between work experience and GMAT score: {correlation_work_exp_gmat}")
```

 Average GMAT Score by Work Industry

```python
avg_gmat_by_industry = df[['work_industry','gmat']].groupby('work_industry').mean()
print(avg_gmat_by_industry)
```

Gender and Work Industry Distribution: This helps identify any trends in the types of industries males and females tend to work in before applying for an MBA

```python
industry_by_gender = df[['work_industry','gender']].groupby('gender').value_counts()
print("Industry distribution by gender:\n", industry_by_gender)
```

Gender Distribution by Mark Category

```python
gender_dist_by_Mark_cat = df[['Mark_cat','gender']].groupby('gender').value_counts()
print(gender_dist_by_Mark_cat)
```

outlier

```python
for col in df:
  if df[col].dtype=='int' or df[col].dtype=='float':
    print(col.capitalize())
    print(f"Mean:{df[col].mean()}")
    print(f"Median:{df[col].median()}")
    print(f"Minimum of {col}:{df[col].min()}")
    print(f"Maximum of {col}:{df[col].max()}")
    print(f"Variance of {col}:{df[col].var()}")
    print(f"Standard deviation:{df[col].std()}")
    print()
    plt.figure(figsize=(8,5))
    plt.boxplot(df[col]) # for detecting outliers, uses numerical values
    plt.title(col)
    plt.show()
    print("_________________________________________________________________________________________________________________")
    print()
```
